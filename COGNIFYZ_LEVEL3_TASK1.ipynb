### Here i don’t have a "Review" column in this dataset but instead have "Rating Text" categories (such as Excellent, Very Good, Good). So i’ll focus on the "Rating Text" column to:

    1. Identify the most common positive and negative keywords related to restaurant ratings (based on the "Rating text").

    2. Calculate the average length of the "Rating Text" and explore if there’s a correlation between the length of rating text and numerical ratings.

### Since there are no actual text reviews in dataset, the Rating Text column represents categorical assessments of restaurants, and  can treat the task as analyzing those categories to identify patterns.

## Map the Rating Text to Positive and Negative Sentiment

#### Rating Text categories (Excellent, Very Good, Good) as positive and negative keywords. For example:Positive Sentiment: Excellent, Very Good. Negative Sentiment: Good (or others if present)

```python
rating_sentiment = {
    'Excellent': 'Positive',
    'Very Good': 'Positive',
    'Good': 'Negative',
    'Average': 'Negative',  # Add more categories if needed
}

# Apply sentiment classification to Rating Text
data['Sentiment'] = data['Rating text'].map(rating_sentiment)

# Check the first few rows
data[['Rating text', 'Sentiment']].head()
```

## Explore Rating Text Length

```python
# Calculate the length of the Rating Text (number of characters)
data['Rating Text Length'] = data['Rating text'].apply(lambda x: len(x.split()))

# Calculate the average length of Rating Text
average_length = data['Rating Text Length'].mean()
print(f"Average length of Rating Text: {average_length:.2f} words")
```

## Investigate the Relationship Between Text Length and Numerical Rating

```python
# Scatter plot to check the relationship between Rating Text Length and Aggregate Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Rating Text Length'], y=data['Aggregate rating'], alpha=0.6)

plt.title('Relationship Between Rating Text Length and Aggregate Rating')
plt.xlabel('Rating Text Length (in words)')
plt.ylabel('Aggregate Rating')
plt.grid(True)
plt.show()

# Calculate the correlation between Rating Text Length and Aggregate Rating
correlation = data[['Rating Text Length', 'Aggregate rating']].corr().iloc[0, 1]
print(f"Correlation between Rating Text Length and Aggregate Rating: {correlation:.2f}")
```

## Frequency of Positive and Negative Sentiments

```python
# Count the number of Positive and Negative sentiments
sentiment_counts = data['Sentiment'].value_counts()

# Plot the sentiment distribution
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='coolwarm')

plt.title('Distribution of Positive and Negative Sentiments in Rating Text')
plt.xlabel('Sentiment')
plt.ylabel('Number of Restaurants')
plt.show()
```

## Identify Common Positive and Negative Keywords (Based on Rating Text)

```python
# Count the most common sentiment (positive vs negative)
rating_sentiment_counts = data['Sentiment'].value_counts()

# Plot the most common sentiments
plt.figure(figsize=(8, 6))
sns.barplot(x=rating_sentiment_counts.index, y=rating_sentiment_counts.values, palette='Blues')

plt.title('Most Common Sentiments (Positive vs Negative)')
plt.xlabel('Sentiment')
plt.ylabel('Number of Restaurants')
plt.show()
```

1. Average Length of Rating Text: 1.34 words

  - The average length of the "Rating Text" is very short, averaging only 1.34 words. This indicates that the ratings in your dataset are likely brief, consisting primarily of single-word descriptors (e.g., "Excellent," "Very Good," "Good"). This could be a result of how restaurants collect feedback or how the data is presented.

2. Correlation between Rating Text Length and Aggregate Rating: -0.60

  - The negative correlation of -0.60 suggests a moderate to strong inverse relationship between the length of the Rating Text and the aggregate rating.

      -- This implies that restaurants with shorter rating texts (such as “Excellent” or “Good”) tend to have higher aggregate ratings, while those with longer or more elaborate text ratings tend to have lower aggregate ratings.

      --This could be a reflection of how customer feedback is structured: shorter comments (like "Excellent") might be associated with highly satisfied customers, while more elaborate ratings may reflect more detailed or neutral feedback, possibly from customers with a slightly lower satisfaction level.

  - Interpretation: It’s important to note that shorter reviews (one-word ratings like "Excellent" or "Very Good") tend to be associated with higher ratings, while longer reviews might indicate more nuanced feedback. Customers giving restaurants high ratings may not feel the need to elaborate, while those giving a lower rating may be more detailed or critical in their comments.
